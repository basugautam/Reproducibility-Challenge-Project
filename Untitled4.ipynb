```python
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
import matplotlib.pyplot as plt
from torch.utils.data import DataLoader, TensorDataset

# Step 1: Generate Synthetic Data
def generate_data(seq_length=50, num_samples=1000):
    X = np.linspace(0, 100, num_samples)
    Y = np.sin(X) + np.random.normal(scale=0.1, size=num_samples)  # Sample time series
    return X, Y

X, Y = generate_data()

# Step 2: Prepare Data for Training
seq_length = 10
def create_sequences(Y, seq_length):
    X_seq, Y_seq = [], []
    for i in range(len(Y) - seq_length):
        X_seq.append(Y[i:i+seq_length])
        Y_seq.append(Y[i+seq_length])
    return np.array(X_seq), np.array(Y_seq)

X_seq, Y_seq = create_sequences(Y, seq_length)
X_train = torch.tensor(X_seq, dtype=torch.float32)
Y_train = torch.tensor(Y_seq, dtype=torch.float32).unsqueeze(1)
train_dataset = TensorDataset(X_train, Y_train)
train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)

# Step 3: Define LSTM Model with Loss Constraints
class LSTMForecaster(nn.Module):
    def __init__(self, input_size, hidden_size, output_size, num_layers=2):
        super(LSTMForecaster, self).__init__()
        self.lstm = nn.LSTM(input_size, hidden_size, num_layers, batch_first=True)
        self.fc = nn.Linear(hidden_size, output_size)

    def forward(self, x):
        _, (hn, _) = self.lstm(x)
        return self.fc(hn[-1])

model = LSTMForecaster(input_size=seq_length, hidden_size=50, output_size=1)
criterion = nn.MSELoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)

# Step 4: Train Model with Loss Shaping Constraints
num_epochs = 50
lambda_k = 0.5  # Constraint hyperparameter

for epoch in range(num_epochs):
    total_loss = 0
    for inputs, targets in train_loader:
        optimizer.zero_grad()
        outputs = model(inputs)
        loss = criterion(outputs, targets)
        constrained_loss = loss + lambda_k * torch.relu(loss - 0.1)  # Loss shaping
        constrained_loss.backward()
        optimizer.step()
        total_loss += loss.item()
    print(f"Epoch {epoch+1}, Loss: {total_loss / len(train_loader):.4f}")

# Step 5: Visualize Results
plt.plot(Y, label='Original Data')
predictions = model(X_train).detach().numpy()
plt.plot(range(seq_length, seq_length + len(predictions)), predictions, label='Predictions')
plt.legend()
plt.show()
```
